```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
NUM_RECTANGLES_PER_TYPE = 15
DIM_1 = 12
DIM_2 = 40
for image in range(100):
    fig = plt.figure(figsize=(16, 12))
    ax = plt.gca()
    ax.axvline(0.5, ls="-", c='w')
    occlusion = np.zeros((1000, 1000), dtype=bool)
    # Background
    rect = patches.Rectangle((0, 0), 1, 1, edgecolor='none', facecolor='k')
    ax.add_patch(rect)
    # Vertical blue bars
    for i in range(NUM_RECTANGLES_PER_TYPE):
        # Generate random positions that do not intersect other rectangles
        while True:
            x, y = np.random.randint(10, 950), np.random.randint(10, 950)
            if 450 < x < 550:
                continue
            if not occlusion[x:x+DIM_1, y:y+DIM_2].any():
                occlusion[x:x+DIM_1, y:y+DIM_2] = True
                break
        rect = patches.Rectangle((0.001*x, 0.001*y), 0.001*DIM_1, 0.001*DIM_2, edgecolor='none', facecolor='c')
        ax.add_patch(rect)
    # Horizontal blue bars
    for i in range(NUM_RECTANGLES_PER_TYPE):
        # Generate random positions that do not intersect other rectangles
        while True:
            x, y = np.random.randint(10, 950), np.random.randint(10, 950)
            if 450 < x < 550:
                continue
            if not occlusion[x:x+DIM_2, y:y+DIM_1].any():
                occlusion[x:x+DIM_2, y:y+DIM_1] = True
                break
        rect = patches.Rectangle((0.001*x, 0.001*y), 0.001*DIM_2, 0.001*DIM_1, edgecolor='none', facecolor='c')
        ax.add_patch(rect)
    # Horizontal red bars
    for i in range(NUM_RECTANGLES_PER_TYPE):
        # Generate random positions that do not intersect other rectangles
        while True:
            x, y = np.random.randint(10, 950), np.random.randint(10, 950)
            if 450 < x < 550:
                continue
            if not occlusion[x:x+DIM_2, y:y+DIM_1].any():
                occlusion[x:x+DIM_2, y:y+DIM_1] = True
                break
        rect = patches.Rectangle((0.001*x, 0.001*y), 0.001*DIM_2, 0.001*DIM_1, edgecolor='none', facecolor='r')
        ax.add_patch(rect)
    # Target bar
    # Generate random positions that do not intersect other rectangles
    while True:
        x, y = np.random.randint(10, 950), np.random.randint(10, 950)
        if 450 < x < 550:
            continue
        if not occlusion[x:x+DIM_1, y:y+DIM_2].any():
            occlusion[x:x+DIM_1, y:y+DIM_2] = True
            break
    rect = patches.Rectangle((0.001*x, 0.001*y), 0.001*DIM_1, 0.001*DIM_2, edgecolor='none', facecolor='r')
    ax.add_patch(rect)
    plt.axis('off')
    plt.subplots_adjust(0,0,1,1,0,0)
    plt.savefig('%d.png' % image)
    #plt.show()
```
